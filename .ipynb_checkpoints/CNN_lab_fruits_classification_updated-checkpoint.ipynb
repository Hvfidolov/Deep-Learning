### Classification de CNN sur BD Fruits

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

### Step 1: Loading Dataset

```python
## variables
img_height, img_width = 32, 32
batch_size = 20

## data loading from file in path
## batch size (taille du lot sur lequel il s'entraine avant de passer)
dataset_train = tf.keras.utils.image_dataset_from_directory("fruits_updated/train", image_size = (img_height, img_width), batch_size = batch_size)
dataset_validation = tf.keras.utils.image_dataset_from_directory("fruits_updated/train", image_size = (img_height, img_width), batch_size = batch_size)
dataset_test = tf.keras.utils.image_dataset_from_directory("fruits_updated/train", image_size = (img_height, img_width), batch_size = batch_size)
```

### Step 2: Visualisation

```python

class_names=['apple','banana','orange','mango','pineapple']
```

### Step 3: Creation du model

```python
model = tf.keras.Sequential([
    tf.keras.layers.Rescaling(1./255),
    
    tf.keras.layers.Conv2D(32, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(64, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(128, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Flatten(),

    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
# Mise  jour de la dernire couche pour inclure les nouvelles classes
model.add(tf.keras.layers.Dense(5, activation='softmax'))  # 5 classes au total
```

### Step 4: Parametrage du modele

```python
model.compile(optimizer='adam', loss='SparseCategoricalCrossentropy', metrics=['accuracy'])
```

### Step 5: Entrainement du modele

```python
model.fit(dataset_train, validation_data = dataset_validation, epochs = 10)
```

### Step 6: Tester le modele

```python
model.evaluate(dataset_test, verbose = 2)
```

```python
model.save('CNN_lab_fruits_classification.h5')
```
